## Run Analysis

```python
if REFIT:
    n_sims = 200


    gen_param_names = ['alpha', 'beta_stage2', 'beta_MB','beta_MF1','eta','cvar_alpha']
    rec_param_names = ['alpha', 'beta_stage2', 'beta_MB','beta_MF1','beta_sticky']

    rows = []
    for sim in range(n_sims):
        print('simulation ...'+str(sim))

        # choose parameters 
        params = [np.random.uniform(low=0.05,high=0.6),
                  np.random.uniform(low=10,high=20),
                  np.random.uniform(low=10,high=20),
                  np.random.uniform(low=10,high=20),
                  np.random.uniform(low=0.001,high=0.09),
                  np.random.uniform(low=0.1,high=0.99)]
        params_t = [transform_params(param,param_name) for param,param_name in zip(params,gen_param_names)]


        # generate data from cvar model 
        generative_results = model(params_t,
                                   model_name='dcvar_mb_mf',
                                  stage1_choices=None,
                                  stage2_choices=None,
                                  stage2_states=None,
                                  outcomes=None,
                                  return_nllk=False,
                                  generative=True,
                                  drifting_probs=drifting_probs,
                                  seed_per_trial=False)

        # fit data with mean model
        fit_results = fit_model(generative_results,
                                  model_name='mean_mb_mf_sticky',
                                  multi_starts=4,
                                  parrallel=True)

        # store row
        gen = {'gen_'+param_name:param for param_name, param in zip(gen_param_names,params)}
        rec = {'rec_'+param_name:param for param_name, param in zip(rec_param_names,fit_results['params'])}
        extra = {'sim':sim}
        params_dict = {**gen,**rec,**extra}
        rows.append(params_dict) 

    params_df = pd.DataFrame(rows)
```
