## (Panel a) Example Distributions

```python
plt.figure(figsize=(4,3),dpi=300)
fs = 12

# one distribution 
ps = np.random.normal(loc=0.7,scale=0.1,size=500)
tmp = plt.hist(ps,alpha=0.5,bins=np.arange(0,1,0.05),ec='k',
               color=sns.color_palette()[0],density=True)

# the other
ps = np.random.normal(loc=0.5,scale=0.2,size=500)
plt.hist(ps,alpha=0.5,bins=np.arange(0,1,0.05),ec='k',
         color=sns.color_palette()[4],density=True)

plt.xlim([0,1])

plt.xlabel(r'reward probability',fontsize=fs)
#plt.xlabel(r'reward probability',fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks([],fontsize=fs)
#set_visible(False)

sns.despine(left=True)
```

## (Panel b) Drifting Reward Probabilities 

```python
# get participants data
directory = '../data/all_twostep_data_study1/'

ID = '3E47SOBEYQWRSDTOFVU29QLS1S8ICY'
filename = ID+'.csv'
trial_data = get_data(directory,filename)
drifting_probs=trial_data[['drift 1','drift 2','drift 3','drift 4']].values.astype('float')
```

```python
plt.figure(figsize=(6,3),dpi=300)
fs = 12

for i,ii in zip(range(4),[0,4,1,3]):
    plt.plot(drifting_probs[:,i],color=sns.color_palette()[ii]);
sns.despine()
plt.xlabel

plt.xlabel('trials',fontsize=fs)    
plt.ylabel('reward probability',fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
```
