```python
y_set_for_plotting = [0,0.05,0.1,0.3,0.6,1]
y_set = y_set_for_plotting
```

```python
def myround(x, prec=1, base=.5):
    return round(base * round(float(x)/base),prec)
```

## (Panel a) Single Distribution

```python
np.random.seed(4)
fontsize = 6
fig = plt.figure(figsize=(2.5,2),dpi=500)#facecolor='blue')

alpha=0.1355
lw=0.2

zis_full = np.random.normal(loc=0,scale=1,size=1000)

bins = np.arange(-3,3,0.25)

var,cvar= calc_cvar_from_samples(zis_full,alpha)

bins = np.arange(-3,3,0.25)

tmp = plt.hist(zis_full[zis_full<=var],
         density=False,
               color=sns.color_palette()[0],
               alpha=1,edgecolor='k',
               bins=bins,
               linewidth=lw);

plt.hist(zis_full[zis_full>var],
     density=False,color=sns.color_palette()[0],
         alpha=0.3,
         bins=bins,
         edgecolor='k',
         linewidth=lw);

plt.axvline(x=var,color=sns.color_palette()[0],alpha=1.0,linestyle='--',linewidth=1,zorder=3)
plt.axvline(x=cvar,color='k',alpha=1.0,linestyle='--',linewidth=1,zorder=3,ymax=0.85)

plt.text(0.22,0.76,r'CVaR$_{\alpha}$',
                 transform=plt.gcf().transFigure,clip_on=False,
                 rotation=0,fontsize=fontsize)
    

plt.text(0.3,0.87,r'VaR$_{\alpha}$',
                 transform=plt.gcf().transFigure,clip_on=False,
                 rotation=0,fontsize=fontsize,color=sns.color_palette()[0])
    

plt.xticks(np.arange(-3,3.1,1),fontsize=fontsize)
plt.xlabel(r'outcomes $(z)$',fontsize=fontsize)
plt.yticks([])
sns.despine(left=True)
```

## (Panel b) Stacked Distributions

```python
zis = np.array([-4,-3,-2,-1,
       0,
       1,2,3])
pis = np.array([0.05,0.05,0.1,0.15,
       0.3,
       0.2,0.10,0.05])
np.sum(pis)
```

```python
zis_full = []
for pi,zi in zip(pis,zis):
    ni = pi*100
    zis_full.extend(list(np.repeat(zi,int(ni)).flatten()))
zis_full=np.array(zis_full)
```

```python
fig = plt.figure(figsize=(2.5,2),dpi=500)#facecolor='blue')

fontsize = 6

yis = [5,4,3,2,1]
axes = []
L = len(yis)

for i in range(len(yis)):
    x0 = 0.25 # start slightly in 
    y0 = (L-i)*(1/(L+1))#*0.75 # reverse, then 1/number of items +1 for legend 
    height = (1/L)*0.7
    width = 0.7
    rect = [x0, y0,width,height]
    axes.append(plt.axes(rect))
    

for i,yi in enumerate(yis):
    
    plt.sca(axes[i])
    var,cvar= calc_cvar_from_samples(zis_full,y_set[yi])
    var = myround(var,prec=1,base=1)
    
    bins = np.arange(-5.5,4.5,1)

    plt.hist(zis_full[zis_full<=var],bins=bins,
             density=False,color=sns.color_palette()[0],alpha=1,edgecolor='k',linewidth=0.4);
    if yi!=20:
        plt.hist(zis_full[zis_full>var],bins=bins,
             density=False,color=sns.color_palette()[0],alpha=0.3,edgecolor='k',linewidth=0.4);
    
    plt.axvline(x=cvar,color='k',linestyle='--',linewidth=1,zorder=3)
    
    plt.text(0.15,0.02+(L-i)*(1/(L+1)),r''+str(np.round(y_set_for_plotting[yi],2)),
             transform=plt.gcf().transFigure,clip_on=False,fontsize=fontsize) 

    # X-axis 
    sns.despine(left=True)
    plt.yticks([])
    if i!=len(yis)-1:
        plt.xticks([])
    elif i==len(yis)-1:
        plt.xticks(zis,fontsize=fontsize)
        plt.xlabel(r'outcomes $(z)$',fontsize=fontsize)
    plt.xlim([-5,4])

    # arrow 
    x0=0.1; y0=1/L*0.8
    x1=0.1; y1=0.9#*(L-1)/L
    plt.arrow(x1,y1,x0-x1,y0-y1,transform=plt.gcf().transFigure,
             color='k',clip_on=False,head_width=.01,linewidth=.5)
    plt.arrow(x0,y0,x1-x0,y1-y0,transform=plt.gcf().transFigure,
             color='k',clip_on=False,head_width=.01,linewidth=.5)
    
    plt.text(x0-0.05,y0+0.25*(y1-y0),r'risk preference ($\alpha$)',
                 transform=plt.gcf().transFigure,clip_on=False,
                 rotation=90,fontsize=fontsize)
    
    # ylim 
    plt.ylim([0,45])
    
```
